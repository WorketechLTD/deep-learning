### 2.2.1 Data Loader
The functions build a TensorFlow Dataset object for a specified subset of the dataset (train, val, or test). The functions use the tf.keras.preprocessing.image_dataset_from_directory() method to create the dataset from the directory of images. The validation_split parameter specifies the fraction of the data to use for validation, the subset parameter specifies which subset of the data to use (train, val, or test), the label_mode parameter specifies the type of labels to use (categorical in this case), the seed parameter sets the random seed for reproducibility, the image_size parameter sets the size of the images, and the batch_size parameter sets the batch size for the dataset.

## 2.7 Preprocessing: Normalization & Data Augmentation
This section defines a data augmentation pipeline and normalization function for the training, validation and evaluation datasets. The data_augmentation pipeline includes four operations: RandomFlip, RandomZoom, RandomContrast, and GaussianNoise. The normalize function normalizes the pixel values of the image to the range between 0 and 1. The augment function applies data augmentation to the image by first normalizing the pixel values to the range between 0 and 1, and then applying the data_augmentation pipeline to the image. The train_ds_raw dataset applies only normalization to the training images, while the train_ds_aug dataset applies both normalization and data augmentation to the training images. The val_ds dataset applies only normalization to the validation images.

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip("horizontal_and_vertical"),
  tf.keras.layers.RandomZoom(0.2),
  tf.keras.layers.RandomContrast(0.2),
  tf.keras.layers.GaussianNoise(stddev=0.1)
])

data_normalization = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1. / 255)
])

def normalize(image, label):
  image = tf.cast(image, tf.float32)
  image = data_normalization(image)
  return image, label

def augment(image, label):
  image = tf.cast(image, tf.float32)
  image = image / 255.0 # normalize pixel values between 0 and 1
  image = data_augmentation(image) # apply data augmentation
  return image, label

# Map the augmentation function to the images on the fly
train_ds_raw = train_ds.map(normalize, num_parallel_calls=tf.data.AUTOTUNE)
train_ds_aug = train_ds.map(augment, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.map(normalize, num_parallel_calls=tf.data.AUTOTUNE)
eval_ds = eval_ds.map(normalize, num_parallel_calls=tf.data.AUTOTUNE)
```

### 3.1.1 Basic CNN trained on Non-Augmented Data: Baseline Model

##### Architecture & Training
This code defines a function called basic_cnn_raw that creates a basic convolutional neural network (CNN) model for image classification. This model will be trained on the dataset that is not augmented. The model has multiple convolutional layers with different numbers of filters and kernel sizes, followed by max pooling layers and dropout layers for regularization. The output of the last convolutional layer is flattened and passed through a dense layer with 10 units and a softmax activation function, which outputs a probability distribution over the 10 possible classes. The model is compiled with the Adam optimizer and categorical cross-entropy loss, and the accuracy metric is also monitored during training.

After defining the function, the code calls the function and assigns the resulting model object to the variable basic_cnn1. The model summary and visualization are also generated and saved to a file.

```python
def basic_cnn_raw():
  model = Sequential()
  model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(224,224,3)))
  model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(.5))

  model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
  model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(.5))

  model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
  model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(.5))

  model.add(Flatten())
  model.add(Dense(10, activation='softmax'))

  optimizer = Adam(learning_rate=0.0007)
  model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
  model.summary()
  return model

basic_cnn1 = basic_cnn_raw()
```

### 3.1.2 MobileNetV3 trained on Non-Augmented Data

##### Architecture & Training
This code defines a mobilenet_model_raw() function with a MobileNet V3 model for image classification that will be trained on non-augmented data. It has the following architecture:

* An input layer that accepts images with shape (224, 224, 3)
* A feature extraction layer based on the pre-trained MobileNet V3 model from TensorFlow Hub
* A dropout layer to reduce overfitting
* A fully connected layer with 10 output units (one for each class)
* A softmax activation layer to convert the output logits to class probabilities
* The function then compiles the model with the Adam optimizer, a learning rate of 0.005, and the categorical cross-entropy loss function.
* It also prints a summary of the model architecture and saves a visualization of the model to a file.
* Finally, the function returns the compiled model.

```python
def mobilenet_model_raw():
  mobilenetmodel = "https://tfhub.dev/google/imagenet/mobilenet_v3_small_075_224/feature_vector/5"
  print("Building model with", mobilenetmodel)
  model = tf.keras.Sequential([
      tf.keras.layers.InputLayer(input_shape=IMAGE_SIZE + (3,)),
      hub.KerasLayer(mobilenetmodel, trainable=False),
      tf.keras.layers.Dropout(rate=0.2),
      tf.keras.layers.Dense(10, kernel_regularizer=tf.keras.regularizers.l2(0.0001),
                            activation="softmax")])
  model.build((None,)+IMAGE_SIZE+(3,))
  # Optimizer and learning rate
  adam = tf.keras.optimizers.Adam(learning_rate=0.005)
  # Compile model
  model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
  model.summary()
  return model

mobilenet1 = mobilenet_model_raw()
```

### 3.2.1 Basic CNN trained on augmented data

This code defines a function called basic_cnn_aug that creates a basic convolutional neural network (CNN) model for image classification. This model will be trained on the dataset that is augmented. The model has multiple convolutional layers with different numbers of filters and kernel sizes, followed by max pooling layers and dropout layers for regularization. The output of the last convolutional layer is flattened and passed through a dense layer with 10 units and a softmax activation function, which outputs a probability distribution over the 10 possible classes. The model is compiled with the Adam optimizer and categorical cross-entropy loss, and the accuracy metric is also monitored during training.

After defining the function, the code calls the function and assigns the resulting model object to the variable basic_cnn1. The model summary and visualization are also generated and saved to a file.

##### Architecture & Training

```python
def basic_cnn_aug():
  model = Sequential()
  model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(224,224,3)))
  model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(.5))

  model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
  model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(.5))

  model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
  model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(.5))

  model.add(Flatten())
  model.add(Dense(10, activation='softmax'))

  optimizer = Adam(learning_rate=0.0007)
  model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
  model.summary()
  return model

basic_cnn2 = basic_cnn_aug()
```

### 3.2.2 MobileNetV3 trained on augmented data

##### Architecture & Training
This code defines a mobilenet_model_aug() function with a MobileNet V3 model for image classification that will be trained on augmented data. It has the following architecture:

* An input layer that accepts images with shape (224, 224, 3)
* A feature extraction layer based on the pre-trained MobileNet V3 model from TensorFlow Hub
* A dropout layer to reduce overfitting
* A fully connected layer with 10 output units (one for each class)
* A softmax activation layer to convert the output logits to class probabilities
* The function then compiles the model with the Adam optimizer, a learning rate of 0.005, and the categorical cross-entropy loss function.
* It also prints a summary of the model architecture and saves a visualization of the model to a file.
* Finally, the function returns the compiled model.

```python
def mobilenet_model_aug():
  mobilenetmodel = "https://tfhub.dev/google/imagenet/mobilenet_v3_small_075_224/feature_vector/5"
  print("Building model with", mobilenetmodel)
  model = tf.keras.Sequential([
      tf.keras.layers.InputLayer(input_shape=IMAGE_SIZE + (3,)),
      hub.KerasLayer(mobilenetmodel, trainable=False),
      tf.keras.layers.Dropout(rate=0.2),
      tf.keras.layers.Dense(10, kernel_regularizer=tf.keras.regularizers.l2(0.0001),
                            activation="softmax")])
  model.build((None,)+IMAGE_SIZE+(3,))
  # Optimizer and learning rate
  adam = tf.keras.optimizers.Adam(learning_rate=0.005)
  # Compile model
  model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
  model.summary()
  return model

mobilenet2 = mobilenet_model_aug()
```
